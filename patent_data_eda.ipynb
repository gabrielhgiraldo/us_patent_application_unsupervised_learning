## TFIDF

```python
tf = text.TfidfVectorizer(stop_words=stop_words, ngram_range=(1,2))
```

```python
tfidf_vecs = tf.fit_transform(patent_titles)
```

```python
df = pd.DataFrame(tfidf_vecs.toarray(), columns=tf.get_feature_names())

len(tf.get_feature_names())
```

```python
df.sum().sort_values(ascending=False)
```

## Topic Modeling

```python
def print_topics(topics):
    for i,topic in enumerate(topics):
        print("Topic {}".format(i))
        print(", ".join(topic)+"\n")
```

## NMF

```python
from sklearn.decomposition import NMF
nmf = NMF(n_components=20)
nmf_vecs = nmf.fit_transform(tfidf_vecs)
```

```python
nmf.components_.shape
```

```python
len(df.columns)
```

### topics

```python
features = tf.get_feature_names()
```

```python
topic_words = [[features[idx] for idx in np.flip(np.argsort(component)[-10:])] for component in nmf.components_]
```

```python
print_topics(topic_words)
```
